# 2. Categorical Visualizations


### 1. Plot a countplot for sex using Seaborn.


```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
sb.countplot(data = df , x = "sex")
plt.title('Count of Each Sex')
plt.xlabel('Sex')
plt.ylabel('Count')

# Show the plot
plt.show()
```

### 2. Plot a countplot for smoker using Seaborn.


```python
sb.countplot(data = df , x = "smoker")
plt.title('Total Number of Smokers')
plt.xlabel('Smokers')
plt.ylabel('Count')

# Show the plot
plt.show()
```

### 3. Plot a countplot for day using Seaborn.


```python
sb.countplot(data = df , x = "day", color = "black")
plt.title('Total Number of Days')
plt.xlabel('Days')
plt.ylabel('Frequency')

# Show the plot
plt.show()

```

### 4. Plot a bar chart showing average tip for each day using Matplotlib.


```python
avg_tip = df.groupby('day')['tip'].mean()  #groupby day

plt.figure(figsize=(8,5))
plt.bar(avg_tip.index, avg_tip.values, color='green')

# Add labels and title
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.title('Average Tip per Day')
plt.show()
```

### 5. Plot a bar chart showing average tip for sex using Matplotlib.


```python
avr_tip = df.groupby('sex')['tip'].mean()  #groupby sex

plt.figure(figsize=(8,5))
plt.bar(avr_tip.index, avr_tip.values, color='green')

# Add labels and title
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.title('Average Tip per Day')
plt.show()
```

### 6. Plot a stacked bar chart for day vs smoker to show the number of smokers and non-smokers per day using Matplotlib.


```python
smoker_counts = df.groupby(['day', 'smoker']).size().unstack(fill_value=0)

# Plot stacked bar chart
smoker_counts.plot(kind='bar', stacked=True, figsize=(8,5), color=['skyblue', 'yellow'])

# Add labels and title
plt.xlabel('Day')
plt.ylabel('Number of People')
plt.title('Number of Smokers and Non-Smokers per Day')
plt.xticks(rotation=0)  # Keep day labels horizontal
plt.legend(title='Smoker')

# Show the plot
plt.show()
```

### 7. Plot a violin plot for tip grouped by sex using Seaborn.


```python
plt.figure(figsize=(11,9))
sb.violinplot(data=df, x='sex', y='tip', color="orange")

# Add title and labels
plt.title('Distribution of Tips by Sex')
plt.xlabel('Sex')
plt.ylabel('Tip')
```

**Semantic Similarity Scoring**

```python
from numpy.linalg import norm
```

```python
def cos_sim(A,B):
  return np.inner(A,B)/(norm(A)*norm(B))

def euclidean(A,B):
  return [norm(A-B)]
```

```python
def is_it_sim(textA, textB, thresh=0.28, sim_func = cos_sim):
  embeddingsA = embed([textA])
  embeddingsB =embed([textB])
  sim_score = sim_func(embeddingsA, embeddingsB)
  if sim_score > thresh:
    print("They are SIMILAR!")
  else:
    print("Not at all...")
  return sim_score,sim_score>thresh
```

***Building Text Classification Using Embeddings and GUSE***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
import tensorflow_hub as hub
import tensorflow.keras as keras
import tf_keras

import tensorflow_datasets as tfds
```

```python
### we can set the train and test that we want in tensorflow datraset
```

Import Film Review dataset from TF dataset

```python
train_data, validation_data, test_data = tfds.load(
    name="imdb_reviews",
    split=('train[:60%]','train[60%:]','test'),
    as_supervised =True
)
```

```python
train_examples_batch, train_labels_batch = next(iter(train_data.batch(10)))
```

```python
train_examples_batch
```

```python
train_labels_batch
```

## Load Hub data and create a hub Layer

```python
embedding_model_url = "https://tfhub.dev/google/universal-sentence-encoder/4"
```

```python
hub_layer = hub.KerasLayer(embedding_model_url, input_shape=[], dtype=tf.string, trainable=False)
```

## Build A Neural Net Text Classifier

```python
nlp_model = tf_keras.Sequential()
```

```python
nlp_model.add(hub_layer)
nlp_model.add(tf_keras.layers.Dense(256, activation ="relu"))
#add two dropout layers, just optional
nlp_model.add(tf_keras.layers.Dropout(0.1))

nlp_model.add(tf_keras.layers.Dense(128, activation ="relu"))
nlp_model.add(tf_keras.layers.Dropout(0.1))

nlp_model.add(tf_keras.layers.Dense(64, activation ="relu"))

nlp_model.add(tf_keras.layers.Dense(1, activation = "sigmoid"))    # adding sigmoid here is not nec needed
```

```python
nlp_model.summary()
```

## Compile ansd Train NLP

```python
nlp_model.compile(optimizer = tf_keras.optimizers.Adam(), loss=tf_keras.losses.BinaryCrossentropy(from_logits = False), metrics =["binary_accuracy"])
```

```python
history = nlp_model.fit(train_data.batch(512), epochs =5, validation_data=validation_data.batch(512), verbose =2)
```

### Model Evaluation

```python
results = nlp_model.evaluate(test_data.batch(512), verbose =2)
```

### Inference Function

```python
def get_sentiment(text, model=nlp_model, thresh=0.5):
  p_hat =model.predict([text])[0][0]
  out = (p_hat>thresh).astype('int32')

  print("Vievwer Comment:\n"+text+"\n\n The Review was:")
  if out: print(" It was Good!")
  else: print("Bad Movie!")
  return p_hat
```
